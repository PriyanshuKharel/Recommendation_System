```python
# Impute missing values
df_imputed = df.fillna({'Description': 'Unknown', 'CustomerID': -1})
```

```python
# Convert 'InvoiceDate' to datetime
df_imputed['InvoiceDate'] = pd.to_datetime(df_imputed['InvoiceDate'])

# Display the first few rows of the DataFrame
print("\nDataFrame with 'InvoiceDate' Converted:")
print(df_imputed.head())
```

```python
# Create a DataFrame with customer purchase frequency
customer_frequency = df_imputed.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
customer_frequency.columns = ['CustomerID', 'PurchaseFrequency']

# Display the first few rows
print("\nCustomer Purchase Frequency:")
print(customer_frequency.head())
```

```python
print(df_imputed.groupby('CustomerID')["InvoiceNo"].head())
```

```python
# Create a DataFrame with customer average purchase value
customer_avg_purchase = df_imputed.groupby('CustomerID')['UnitPrice'].mean().reset_index()
customer_avg_purchase.columns = ['CustomerID', 'AvgPurchaseValue']

# Display the first few rows
print("\nCustomer Average Purchase Value:")
print(customer_avg_purchase.head())
```

```python
# Calculate the recency of purchases for each customer
max_date = df_imputed['InvoiceDate'].max()
customer_recency = df_imputed.groupby('CustomerID')['InvoiceDate'].max().reset_index()
customer_recency['Recency'] = (max_date - customer_recency['InvoiceDate']).dt.days

# Display the first few rows
print("\nCustomer Recency of Purchases:")
print(customer_recency[['CustomerID', 'Recency']].head())
```

```python
# Create a DataFrame with the most popular products
popular_products = df_imputed.groupby('StockCode')['Quantity'].sum().reset_index()
popular_products.columns = ['StockCode', 'TotalQuantity']

# Display the first few rows
print("\nMost Popular Products:")
print(popular_products.sort_values(by='TotalQuantity', ascending=False).head())
```

```python
# Create a DataFrame with daily sales
daily_sales = df_imputed.groupby(df_imputed['InvoiceDate'].dt.date)['Quantity'].sum().reset_index()
daily_sales.columns = ['Date', 'TotalQuantity']

# Display the first few rows
print("\nDaily Sales:")
print(daily_sales.tail())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


# Merging the DataFrames
customer_features = pd.merge(customer_frequency, customer_avg_purchase, on='CustomerID', how='left')
customer_features = pd.merge(customer_features, customer_recency[['CustomerID', 'Recency']], on='CustomerID', how='left')

# Filling any potential remaining missing values
customer_features.fillna(0, inplace=True)

# Displaying the first few rows
print("\nCustomer Features:")
print(customer_features.head())

# Selecting relevant features for clustering
X = customer_features[['PurchaseFrequency', 'AvgPurchaseValue', 'Recency']]

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_tr1ansform(X)

# Displaying the standardized features
customer_features_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print("\nStandardized Customer Features:")
print(customer_features_scaled.head())
```

```python
# Display column names in customer_features_scaled
print(customer_features_scaled.columns)
```

```python
# Choose the number of clusters (you can adjust this based on your goals)
num_clusters = 3

# Apply K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
customer_features_scaled['Cluster'] = kmeans.fit_predict(X_scaled)

# Combine the original features with cluster assignments
customer_clusters = pd.concat([customer_features, customer_features_scaled['Cluster']], axis=1)

# Calculate mean values for each cluster
cluster_means = customer_clusters.groupby('Cluster').mean()

# Display the mean values for each cluster
print("\nCluster Means:")
print(cluster_means)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of PurchaseFrequency vs AvgPurchaseValue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_clusters, x='PurchaseFrequency', y='AvgPurchaseValue', hue='Cluster', palette='viridis')
plt.title('Cluster Visualization: PurchaseFrequency vs AvgPurchaseValue')
plt.show()

# Scatter plot of PurchaseFrequency vs Recency
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_clusters, x='PurchaseFrequency', y='Recency', hue='Cluster', palette='viridis')
plt.title('Cluster Visualization: PurchaseFrequency vs Recency')
plt.show()

# Scatter plot of AvgPurchaseValue vs Recency
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_clusters, x='AvgPurchaseValue', y='Recency', hue='Cluster', palette='viridis')
plt.title('Cluster Visualization: AvgPurchaseValue vs Recency')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Concatenate, Dense
from tensorflow.keras.optimizers import Adam

# Sample data
# ... (same as before)

# Encode categorical variables
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()
country_encoder = LabelEncoder()

df['CustomerID'] = user_encoder.fit_transform(df['CustomerID'])
df['StockCode'] = item_encoder.fit_transform(df['StockCode'])
df['Country'] = country_encoder.fit_transform(df['Country'])

# Split the data into training and testing sets
train, test = train_test_split(df, test_size=0.2, random_state=42)

# Define the model architecture
num_users = len(user_encoder.classes_)
num_items = len(item_encoder.classes_)
num_countries = len(country_encoder.classes_)
embedding_size = 50

# User embeddings
user_input = Input(shape=(1,), name='user_input')
user_embedding = Embedding(input_dim=num_users, output_dim=embedding_size, input_length=1)(user_input)
user_flat = Flatten()(user_embedding)

# Item embeddings
item_input = Input(shape=(1,), name='item_input')
item_embedding = Embedding(input_dim=num_items, output_dim=embedding_size, input_length=1)(item_input)
item_flat = Flatten()(item_embedding)

# Country embeddings
country_input = Input(shape=(1,), name='country_input')
country_embedding = Embedding(input_dim=num_countries, output_dim=embedding_size, input_length=1)(country_input)
country_flat = Flatten()(country_embedding)

# Concatenate user, item, and country embeddings
concatenated = Concatenate()([user_flat, item_flat, country_flat])

# Dense layers for prediction
dense1 = Dense(64, activation='relu')(concatenated)
output = Dense(1)(dense1)

# Model
merged = Model(inputs=[user_input, item_input, country_input], outputs=output)

# Compile the model
merged.compile(optimizer=Adam(lr=0.001), loss='mean_squared_error')

# Train the model
history = merged.fit([train['CustomerID'], train['StockCode'], train['Country']], train['Quantity'],
                     batch_size=64, epochs=1000, validation_data=([test['CustomerID'], test['StockCode'], test['Country']], test['Quantity']))

# Predictions
predictions = merged.predict([test['CustomerID'], test['StockCode'], test['Country']])

# Evaluate the model
mse = np.mean((predictions.flatten() - test['Quantity'])**2)
print(f'Mean Squared Error on Test Set: {mse}')
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from implicit.als import AlternatingLeastSquares
from scipy.sparse import coo_matrix

# Sample data
# ... (assuming you have a DataFrame named 'df' with columns like 'CustomerID', 'StockCode', 'Recency', 'AvgPurchaseValue', etc.)

# Encode categorical variables
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()
country_encoder = LabelEncoder()

df['CustomerID'] = user_encoder.fit_transform(df['CustomerID'])
df['StockCode'] = item_encoder.fit_transform(df['StockCode'])
df['Country'] = country_encoder.fit_transform(df['Country'])

# Create a sparse user-item matrix
user_item_matrix = coo_matrix(( (df['CustomerID'], df['StockCode'])))

# Train ALS model
model = AlternatingLeastSquares(factors=50, regularization=0.01, iterations=50)
model.fit(user_item_matrix)

# Recommend similar products for a given StockCode
def recommend_similar_products(stock_code, model, item_encoder, num_recommendations=5):
    item_id = item_encoder.transform([stock_code])[0]
    similar_items = model.similar_items(item_id, N=num_recommendations + 1)[1:]  # Exclude the input item itself
    similar_stock_codes = item_encoder.inverse_transform(similar_items[:, 0])
    return similar_stock_codes

# Example: Recommend similar products for a given StockCode
input_stock_code = '21773'
similar_products = recommend_similar_products(input_stock_code, model, item_encoder)
print(f"Similar products to '{input_stock_code}': {similar_products}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import LabelEncoder

# Sample data
# ... (assuming you have a DataFrame named 'df' with columns like 'CustomerID', 'StockCode', 'Recency', 'AvgPurchaseValue', etc.)

# Encode categorical variables
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()
country_encoder = LabelEncoder()

df_imputed['CustomerID'] = user_encoder.fit_transform(df_imputed['CustomerID'].unique())
df_imputed['StockCode'] = item_encoder.fit_transform(df_imputed['StockCode'])
df_imputed['Country'] = country_encoder.fit_transform(df_imputed['Country'])

# Split the data into training and testing sets
train, test = train_test_split(df_imputed, test_size=0.2, random_state=42)

# Create user-item matrix
user_item_matrix = train.pivot(index='CustomerID', columns='StockCode').fillna(0)

# Train Nearest Neighbors model
knn_model = NearestNeighbors(n_neighbors=5, metric='cosine', algorithm='brute')
knn_model.fit(user_item_matrix.values)

# Recommend similar products for a given StockCode
def recommend_similar_products(stock_code, knn_model, item_encoder, num_recommendations=5):
    item_id = item_encoder.transform([stock_code])[0]
    distances, indices = knn_model.kneighbors(user_item_matrix.iloc[:, item_id].values.reshape(1, -1), n_neighbors=num_recommendations + 1)
    similar_stock_codes = item_encoder.inverse_transform(indices.flatten()[1:])
    return similar_stock_codes

# Example: Recommend similar products for a given StockCode
input_stock_code = '21773'
similar_products = recommend_similar_products(input_stock_code, knn_model, item_encoder)
print(f"Similar products to '{input_stock_code}': {similar_products}")
```
